```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt

edges = pd.read_csv('part1_edges_test.csv')
edges
```

```python
nodes = pd.read_csv('all_nodes_test.csv')
```

```python
nodes_only = nodes [['Id' , 'Label', 'type']].copy()
nodes_only
```

```python
G = nx.from_pandas_edgelist(edges, source='Source', target='Target', edge_attr='Weight', create_using=nx.DiGraph())
```

```python
plt.figure(figsize=(40,40))
node_size=100
nx.draw_networkx(G, with_labels=True, font_size=24, node_size=node_size)
```

```python
from networkx import number_of_nodes
from networkx import number_of_edges

print ("Number of nodes:", number_of_nodes(G))
print ("Number of edges:", number_of_edges(G))
```

```python
#Centrality Measures
from networkx import eigenvector_centrality
from networkx import katz_centrality
from networkx import hits

degreeCent = nx.degree_centrality(G)
outDegree = nx.out_degree_centrality(G)
inDegree = nx.in_degree_centrality(G)

betCent = nx.betweenness_centrality(G, normalized=True, endpoints=True)

closenessCent = nx.closeness_centrality(G, u=None, distance=None, wf_improved=True)

eigCent= eigenvector_centrality(G, max_iter=100, tol=1e-06, nstart=None, weight=None)

katzCent = katz_centrality(G, alpha=0.1, beta=1.0, max_iter=1000, tol=1e-06, nstart=None, normalized=True, weight=None)

hits = hits(G, max_iter=100, tol=1e-08, nstart=None, normalized=True)
```
